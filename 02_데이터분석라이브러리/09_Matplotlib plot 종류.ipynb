```python
import matplotlib.pyplot as plt

# 그래프에 대한 기본설정
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['font.size'] = 16
plt.rcParams['axes.unicode_minus'] = False
```

## Matplotlib plot 종류
### 선 그래프
- 어떠한 데이터의 흐름에 따라 변화되는 데이터를 표현할 때 사용한다.

```python
x = [1, 2, 3]
y = [2, 3, 1]
y2 = [1, 2, 3]

# 그래프의 제목
plt.title('선그래프', fontsize=30, pad=30)

# 선 그래프를 그린다.
plt.plot(x, y, label='선1')
plt.plot(x, y2, label='선2')

# 범례 표시
plt.legend(loc=1, bbox_to_anchor=(1.1, 1.0))

# 각 축의 이름
plt.xlabel('x축', labelpad=30, fontsize=30)
plt.ylabel('y축', rotation=0, labelpad=30, fontsize=30)

#  각 축의 눈금
plt.xticks([1, 2, 3], ['가', '나', '다'], fontsize=30)
plt.yticks([0, 2, 4], ['하', '중', '상'], fontsize=30)

plt.show()
```

### 막대 그래프
- 데이터의 양을 비교하기 위해 사용한다.

```python
x = [1, 2, 3]
y = [2, 3, 1]

# 그래프의 제목
plt.title('막대 그래프', fontsize=30, pad=30)

# 막대 그래프를 그린다.
plt.bar(x, y, label='막대')

# 범례 표시
plt.legend(loc=1, bbox_to_anchor=(1.1, 1.0))

# 각 축의 이름
plt.xlabel('x축', labelpad=30, fontsize=30)
plt.ylabel('y축', rotation=0, labelpad=30, fontsize=30)

#  각 축의 눈금
plt.xticks([1, 2, 3], ['가', '나', '다'], fontsize=30)
plt.yticks([0, 2, 4], ['하', '중', '상'], fontsize=30)

plt.show()
```

### 가로 막대 그래프

```python
x = [1, 2, 3]
y = [2, 3, 1]

# 그래프의 제목
plt.title('막대 그래프', fontsize=30, pad=30)

# 막대 그래프를 그린다.
plt.barh(x, y, label='막대')

# 범례 표시
plt.legend(loc=1, bbox_to_anchor=(1.1, 1.0))

# 각 축의 이름
plt.ylabel('x축', labelpad=30, fontsize=30)
plt.xlabel('y축', rotation=0, labelpad=30, fontsize=30)

#  각 축의 눈금
plt.yticks([1, 2, 3], ['가', '나', '다'], fontsize=30)
plt.xticks([0, 2, 4], ['하', '중', '상'], fontsize=30)

plt.show()
```

### 파이차트
- 카테고리별 값의 상대적인 비교를 할 때 사용한다
- 막대 그래프는 양을 비교하지만 파이는 비율을 비교한다.

```python
# 데이터
values = [15, 30, 40, 10]
# 각 파이의 레이블
labels = ['값1', '값2', '값3', '값4']
# 파이의 색상
# https://matplotlib.org/stable/gallery/color/named_colors.html
# 여기에는 css 색상은 모두 사용할 수 있다
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
# 조각이 튀어나오는 정도(강조)
explode = [0, 0.1, 0, 0]

plt.title('파이차트')

# labels : 각 파이에 표시될 레이블
# colors : 파이의 색상
# explode : 튀어나오는 정도
# startangle : 원의 시작 각도(기본 0도는 3시방향, 시계 반대 방향)
# counterclock : 그리는 방향을 시계 반대 방향으로 할 것인가.. 기본이 True,
# False시 시계 방향으로 그린다.
# autopct : 파이에 비율을 표시할 것인가... '%.2f%%'
plt.pie(values, labels=labels, colors=colors, explode=explode, startangle=90,
        counterclock=False, autopct='%.2f%%')

plt.show()
```

### 히스토그램
- 막대 그래프와 유사해 보이지만 구간별 양을 비교하는 용도로 사용한다.
- 범위형 데이터에 사용

```python
import numpy as np

x = np.random.randn(1000)

plt.title('히스토그램')
# bins : 구간의 개수(생략시 10)
plt.hist(x, bins=30)
plt.show()
```

### 산점도
- 2차원 데이터의 상관 관계를 파악할 수 있다.

```python
# 0 ~ 1 사이의 랜덤값 100개 추출
x1 = np.random.rand(100)
y1 = np.random.rand(100)

x2 = np.random.rand(100)
y2 = np.random.rand(100)

plt.title('산점도')
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
```
