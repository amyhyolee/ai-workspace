## Seaborn (시각화)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 그래프에 대한 기본설정
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['font.size'] = 16
plt.rcParams['axes.unicode_minus'] = False
```

### 1차원 분포 플롯
- 데이터의 분포를 확인하기 위해서는 산점도(scatter)를 사용한다.
- 산점도는 x축과 y축이 지정되어야 하기 때문에 1차원 데이터는 표현할 수 없다.
- 이에 seaborn에서 1차원 데이터의 분포를 확인할 수 있는 것을 제공한다.

```python
iris = pd.read_csv('data/iris.csv')
iris.columns = ['꽃받침길이', '꽃받침너비', '꽃잎길이', '꽃잎너비', '품종']
iris
```

```python
# Rug Plot
# 어떤 값들이 저장되어 있는지 분포를 확인할 수 있다.
sns.rugplot(iris['꽃잎길이'])
plt.title('꽃잎 길이에 대한 Rug Plot')
plt.show()
```

```python
# kde : 히스토그램은 막대그래프로 값 부분을 표현하지만 kde는 선으로 표현한다.
sns.kdeplot(iris['꽃잎길이'])
plt.title('kde plot')
plt.show()
```

```python
# displot : rug, kde, 히스토그램을 같이 표현한다.
sns.displot(iris['꽃잎길이'], kde=True, rug=True)
plt.title('displot')
plt.show()
```

### 카운트 플롯
- 특정 컬럼의 데이터의 개수를 막대 그래프로 표현한 그래프이다.
- 특정 컬럼의 데이터의 각 종류별로 몇 개씩 저장되어 있는지 확인할 수 있다.

```python
sns.countplot(data=iris, x='품종')
plt.title('count plot')
plt.show()
```

### 다차원 데이터
- 데이터 변수가 여러 개인 다차원 데이터를 표현할 수 있다.
- 분석하고자 하는 데이터가 모두 범위 데이터인 경우
- 분석하고자 하는 데이터가 모두 카테고리 값인 경우
- 분석하고자 하는 데이터가 범위형과 카테고리 형이 섞여 있는 경우

### 2차원 실수형 데이터

```python
# joint plot : 산점도와 히스토그램을 같이 표현한다.
sns.jointplot(data=iris, x='꽃받침길이', y='꽃받침너비')
plt.suptitle('joint plot', y=1.02)
plt.show()
```

```python
# 결과데이터를 지정하면 결과데이터의 종류로 구분하여 그려준다.
# hue : 결과데이터로 사용할 컬럼을 지정한다.
sns.jointplot(data=iris, x='꽃받침길이', y='꽃받침너비', hue='품종')
plt.suptitle('joint plot', y=1.02)
plt.show()
```

```python
# kind 에 kde를 넣어주면 kde 밀도 그래프를 그릴 수 있다.
sns.jointplot(data=iris, x='꽃받침길이', y='꽃받침너비', kind='kde')
plt.suptitle('joint plot', y=1.02)
plt.show()
```

```python
sns.jointplot(data=iris, x='꽃받침길이', y='꽃받침너비', kind='kde', hue='품종')
plt.suptitle('joint plot', y=1.02)
plt.show()
```

### PairPlot
- 산점도의 경우에는 2차원 데이터, 즉 컬럼 두 개를 지정하여 표현한다.
- 만약 데이터 프레임의 컬럼이 다 수인 경우 산점도를 그리려면 개발자가 꽤 많은 그래프를 그려야 한다.
- PariPlot은 데이터 프레임의 각 컬럼들을 조합한 산점도를 모두 표현한다.
- 같은 컬럼의 조합에서는 히스토그램이 kde로 표현한다.

```python
sns.pairplot(iris)
plt.suptitle('pair plot', y=1.02)
plt.show()
```

```python
sns.pairplot(iris, hue='품종')
plt.suptitle('pair plot', y=1.02)
plt.show()
```

### heatmap
- 주어진 값이 높거나 낮음을 색상으로 표현하는 그래프
- 주로 컬럼간의 상관관계나 양을 색상으로 표현해 확인하는 그래프
- 직접 통계값을 구해서 그려야 한다.

```python
flights = sns.load_dataset('flights')
flights
```

```python
# pivot 테이블
# 3개의 컬럼을 지정한다.
# 첫 번째 컬럼은 행, 두 번째 컬럼은 열로로 지정하여 세번째 컬럼을 값으로하는
# 데이터들 의미한다.
# 로우 하나의 데이터를 테이블 형태로 만든 테이블
df = flights.pivot('month', 'year', 'passengers')
df
```

```python
sns.heatmap(df)
plt.title('heat map')
plt.show()
```

```python
# 상관관계
# 두 컬럼의 조합으로 값의 변동율을 구한다.
# 양수면 증가, 음수면 감소에 해당하고 1.0 이면 증감이
# 완벽하게 동일하다고 해석하면 된다.
# 증감이 중요한 것이 아닌 변동량이 중요하다.
a1 = iris.corr()
a1
```

```python
# annot : True를 주면 값이 출력된다
# cmap : 색상을 설정한다.
# https://seaborn.pydata.org/tutorial/color_palettes.html
sns.heatmap(a1, annot=True, cmap='summer')
plt.show()
```
