### Matplotlib
- 파이썬에서 데이터를 시각화 할 때 사용하는 대표 라이브러리
- 다른 라이브러리들은 이 라이브러리를 이용해 개선하거나 추가적으로 제공되는 것들이다.

```python
import matplotlib.pyplot as plt

# 그래프에 대한 기본설정
# 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.pcParams['font.family'] = 'AppleGothic'
# 그래프 크기
plt.rcParams['figure.figsize'] = 20, 10
# 글자 크기
plt.rcParams['font.size'] = 16
# 폰트 적용시 음수 기호 깨는 것을 방지한다.
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 기본 그래프
data1 = [1, 4, 9, 16]

plt.plot(data1)

plt.show()
```

```python
# 그래프의 타이틀
data1 = [1, 4, 9, 16]

plt.plot(data1)
# 타이틀
plt.title('기본 그래프')

plt.show()
```

```python
# 데이터를 하나만 정해주면 데이터를 y축에 적용하고
# x 축은 자동으로 설정되거나 오류가 발생한다.

# 데이터를 두 개 설정하면
# 첫 번째 것을 X 축, 두 번째 것을 Y축의 좌표로 사용한다.
data1 = [1, 4, 9, 16]
data2 = [10, 20, 30, 40]

plt.plot(data2, data1)
plt.title('기본 그래프')

plt.show()
```

```python
# 각 축의 이름
data1 = [1, 4, 9, 16]
data2 = [10, 20, 30, 40]

plt.plot(data2, data1)
plt.title('기본 그래프')

# x 축의 이름
plt.xlabel('x 축')
# y 축의 이름
plt.ylabel('y 축')

plt.show()
```

```python
# 사용할 폰트들은 전부 시스템 폰트로 설치가 되어 있어야 한다.
# 폰트파일에서 우클릭 후 설치~!!!

# 개별 파일로 부터 폰트 이름을 알아온다.
prop1 = fm.FontProperties(fname='font/NanumPen.ttf')
name1 = prop1.get_name()

prop2 = fm.FontProperties(fname='font/SingleDay-Regular.ttf')
name2 = prop2.get_name()

prop3 = fm.FontProperties(fname='font/Gugi-Regular.ttf')
name3 = prop3.get_name()

print(name1)
print(name2)
print(name3)
```

```python
# 개별 폰트 지정
# family : 폰트, size : 크기, color : 색상
# 색상 : 
# https://www.w3schools.com/colors/colors_names.asp
# https://www.w3schools.com/colors/colors_picker.asp

option1 = {
    'family' : name1,
    'size' : 24,
    'color' : 'blue'
}

option2 = {
    'family' : name2,
    'size' : 16,
    'color' : 'red'
}

option3 = {
    'family' : name3,
    'size' : 60,
    'color' : '#117ee4'
}

data1 = [1, 4, 9, 16]
data2 = [10, 20, 30, 40]

plt.plot(data2, data1)
plt.title('기본 그래프', fontdict=option1)

plt.xlabel('x 축', fontdict=option2)
plt.ylabel('y 축', fontdict=option3)

plt.show()
```

```python
# 그래프 스타일 지정
data1 = [1, 4, 9, 16]
data2 = [10, 20, 30, 40]

plt.plot(data2, data1, 'rs--')
plt.title('기본 그래프')


plt.show()
```

- 그래프의 스타일
- '색상 마커종류 선종류' 형태로 작성한다(rs--)
- [색상](https://matplotlib.org/stable/gallery/color/named_colors.html)

![색상맵](image/sphx_glr_named_colors.png)
- [마커](https://matplotlib.org/stable/api/markers_api.html)
- 선종류
    - '-' : 실선
    - '--' : dash
    - '-.' : 점과 dash
    - ':' : 점선

```python
# 그래프 스타일 지정
data1 = [1, 4, 9, 16]
data2 = [10, 20, 30, 40]

plt.plot(data2, data1, 'mD-.')
plt.title('기본 그래프')


plt.show()
```

```python
# 상세 설정
# c : 선색상
# lw : 선 굵기
# ls : 선 스타일
# marker : 마커 종류
# ms : 마커 크기
# mec : 마커 색상
# mew : 마커 선 굵기
# mfc : 마커 내부 색상

# 그래프 스타일 지정
data1 = [1, 4, 9, 16]
data2 = [10, 20, 30, 40]

plt.plot(data2, data1, c='r', lw=5, ls=':', marker='o', ms=15
         , mec='g', mew=5, mfc='b')
plt.title('기본 그래프')


plt.show()
```

```python
# 각 축의 범위
data1 = [1, 4, 9, 16]
data2 = [10, 20, 30, 40]

plt.plot(data2, data1)
plt.title('기본 그래프')

# x 축의 범위
plt.xlim(-20, 60)
# y 축의 범위
plt.ylim(-30, 30)


plt.show()
```

```python
# 각 축의 눈금
# 각 축의 눈금은 데이터를 보고 자동으로 설정된다.
# 이 축의 눈금을 바꾸고 싶을 때 사용한다.

data1 = [1, 4, 9, 16]
data2 = [10, 20, 30, 40]

plt.plot(data2, data1)
plt.title('기본 그래프')

# x축 눈금
plt.xticks([10, 20, 30, 40])
# y축 눈금
plt.yticks([0, 5, 10, 15, 20])

plt.show()
```

```python
# 그리드 설정
# 구형 matplotlib는 그리드가 기본으로 나오는 것이였다.
# 각 축의 눈금
# 각 축의 눈금은 데이터를 보고 자동으로 설정된다.
# 이 축의 눈금을 바꾸고 싶을 때 사용한다.

data1 = [1, 4, 9, 16]
data2 = [10, 20, 30, 40]

plt.plot(data2, data1)
plt.title('기본 그래프')

# 그리드 표시
plt.grid()
# 만약 기본으로 그리드가 나오는 구형 버전을 경우에는
# 그리드가 사라지게 하고 싶을 때 사용한다.
# plt.grid(False)

plt.show()
```

```python
# 범례
data1 = [1, 4, 9, 16]
data2 = [10, 20, 30, 40]
data3 = [16, 9, 4, 1]

plt.plot(data2, data1, label='선1')
plt.plot(data2, data3, label='선2')
plt.title('기본 그래프')

# 범례 표시
# 각 그래프의 label 속성의 문자열을 표시한다.
# loc : 범례의 위치, 생략시 0
# 0 : best, 가장 최적의 위치를 자동으로 선정한다.
# (가급적 그래프와 겹치지 않는 위치)
# 1 : 우상단, 2 : 좌상단, 3 : 좌하단, 4 : 우하단, 5 : 중앙우측
# 6 : 중앙좌측, 7 : 중앙우측, 8 : 중앙하단, 9 : 중앙상단, 10 : 정중앙

# bbox_to_anchor : 현재 위치를 기준으로 범례의 위치를 조정한다.(비율)
# (가로위치, 세로위치) : 작아지면 좌측 혹은 위로, 커지만 우측 혹은 아래로
plt.legend(loc=1, bbox_to_anchor=(1.1, 1.0))


plt.show()
```
