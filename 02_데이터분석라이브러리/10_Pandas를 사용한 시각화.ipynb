```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 그래프에 대한 기본설정
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['font.size'] = 16
plt.rcParams['axes.unicode_minus'] = False
```

## Pandas를 사용한 시각화

```python
# 랜덤값 3개가 들어 있는 행렬 100개를 생성한다.
x = np.random.randn(100, 3)
# 2022년 1월 1일 부터 100일치 날짜를 가져온다.
index = pd.date_range('2022/01/01', periods=100)

df1 = pd.DataFrame(x, index=index, columns=['A', 'B', 'C'])
df1
```

### 선 그래프
- 데이터 프레임의 컬럼 하나가 선 하나가 된다.
- index가 x축으로 지정된다.
- 전체 값이 최대 최소 범위가 y축으로 설정된다.
- 컬럼의 이름이 각 선의 이름이 되고 범례에 표시된다.

```python
df1.plot()

plt.title('선 그래프')
plt.xlabel('시간')
plt.ylabel('데이터')

plt.show()
```

### 다른 그래프 그리기
- 데이터 프레임의 plot 함수에 kind라는 매개변수에 그래프의 종류를 설정할 수 있다.
- 그래프의 일부는 함수로 제공되기도 한다.
- bar, pie, hist, kde, box, scatter, area 등..

```python
iris = pd.read_csv('data/iris.csv')
iris.columns =['꽃받침길이', '꽃받침너비', '꽃잎길이', '꽃잎너비', '품종']
iris
```

```python
# Series 하나로 막대 그래프를 그린다.
# 막대가 너무 많아서...^^;;;;
# 인덱스가 x 축, 값이 y축
df1 = iris.head(20)

df1['꽃받침길이'].plot(kind='bar')
plt.title('막대그래프')
plt.show()
```

```python
# 일부는 함수로도 제공된다.
df1 = iris.head(20)

df1['꽃받침길이'].plot.bar()
plt.title('막대그래프')
plt.show()
```

```python
# 그룹 막대 그래프
df1 = iris.drop('품종', axis=1)

df2 = iris.head(20)

df2.plot.bar()

plt.legend(loc=1, bbox_to_anchor=(1.15, 1.0))

plt.show()
```

### 파이 그래프
- Series를 통해 그린다.
- 값 하나가 파이 조각 하나가 된다.

```python
a1 = iris['품종'].value_counts()

# 매개변수들은 앞서 살펴본 matplotlib와 동일하다.
a1.plot.pie(autopct='%.2f%%', startangle=90)

plt.ylabel('품종', rotation=0, labelpad=100)

plt.show()
```

### 히스토그램
- 데이터 프레임이 가지고 있는 범위형 데이터만을 사용해야 한다.

```python
# 컬럼하나(Series)
# bins : 구간의 개수
iris['꽃받침길이'].plot.hist(bins=30)
plt.show()
```

```python
# 다수 컬럼
df1 = iris.drop('품종', axis=1)

# 다수의 컬럼의 그래프가 겹쳐지기 때문에
# alpha 매개변수를 이용해 투명도를 줘야 한다.(0 ~ 1)
df1.plot.hist(bins=30, alpha=0.5)
plt.show()
```

### kde
- 데이터의 밀도를 측정하여 밀도를 시각화 하는 그래프이다.
- 다수의 컬럼을 이용해 히스토그램을 그릴 것을 대신해서 사용할 수 있다.

```python
df1 = iris.drop('품종', axis=1)

df1.plot.kde()
plt.show()
```

### boxplot
- 극단치 확인용으로 사용한다.

```python
# 카테고리형 데이터를 날린다.
df1 = iris.drop('품종', axis=1)

df1.plot.box()
plt.show()
```

```python
df1.boxplot()
plt.show()
```
