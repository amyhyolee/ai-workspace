# 머신러닝

### 머신러닝이란?
- 데이터의 패턴을 컴퓨터에게 학습시키는 개념

### 지도 학습
- 학습시 입력데이터와 결과데이터를 모두 주는 학습 방법
- 지도 학습 알고리즘들은 모두 결과를 예측하는 목적을 가지고 있다.
- 각각의 알고리즘은 학습 방식과 예측 방식이 다르다.
- 속도가 빠른 알고리즘은 성능이 떨어지는 편이고
- 성능이 좋은 알고리즘은 속도가 느린편이다.

### 비지도 학습
- 학습시 입력데이터만 주는 학습 방법
- 비지도 학습은 알고리즘에 따라 목적이 다르므로 알고리즘을 가지고 무엇을 할 수 있는지 파악하는 것이 중요하다.

### 강화학습
- 환경과 움직임, 상벌을 결정하여 상을 받는 쪽의 결과가 나올 수 있는 데이터를 생산하는 학습 방법

### 회귀
- 지도학습의 한 종류로 데이터의 패턴에 따라 수치를 계산하여 예측하는 방식

### 분류
- 지도학습의 안 종류로 데이터의 패턴에 따라 카테고리형 데이터를 예측하는 방식
