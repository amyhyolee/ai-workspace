- 선형모델 : 회귀기법을 이용한 머신러닝 알고리즘. 가장 기본이 되는 알고리즘. 예측선을 찾아 예측하는 방식. 분류에서 예측선은 경계선으로 사용한다.
- 최근접 이웃 : 새롭게 발생된 데이터 주변의 데이터를 보고 에측을 한다. 분류에서는 다수결, 회귀에서는 평균을 구한다.
- 트리 : 주어진 입데이터를 가지고 질문을 생성하여 스무고개 방식으로 예측하는 알고리즘. 과적합이 발생활 확률이 매우 높은 알고리즘이며 회귀에서 성능이 엉망이다.
- 앙상블 : 다수의 알고리즘을 모아 각 알고리즘이 예측하는 결과를 취합하여 최종 결과를 결정하는 방식
- 부스팅 : 앙상블과 비슷하지만 잘못 예측한 결과에 대해 오차를 보정해나가는 방식으로 학습을 한다.
- 보팅 : 다수의 알고리즘을 모아 각 알고리즘이 예측하는 결과를 취합하여 최종 결과를 결정하는 방식. 앙상블과 유사하지만 보팅은 결과 취합 방식과 사용할 알고리즘들을 개발자가 결정할 수 있다.
