### LabelEncoder
- 컬럼내의 값을 숫자로 변환할 때 사용한다.
- 지정된 컬럼안에 들어있는 값의 종류를 파악하여 변환하다.

### Series 사용

```python
df1 = pd.read_csv('data/iris.csv')
y = df1['target']
y
```

```python
# Series에 있는 데이터를 확인해서 숫자로 변환한다.
y2, index = y.factorize()
display(y2)
display(index)
```

- factorize 함수
- Series안에 들어있는 값이 무엇인지 확인하여 숫자로 변환해준다.
- 개발자가 어떤 값들이 들어 있는지 파악하지 않아도 된다.
- 단, 원래의 문자열로 변환(예측된 결과)하기 위해서는 개발자가 직접 구현해 줘야 한다.

### map

```python
dict1 = {
    'setosa' : 0,
    'versicolor' : 1,
    'virginica' : 2
}

y2 = y.map(dict1)
y2
```

- map 함수
- 딕셔너리 안에 어떠한 값을 무엇으로 바꿀지를 직접 지정할 수 있다.
- 만약 딕셔너리안에 명시하지 않은 데이터가 있다면 결측치가 되기 때문제 주의해야 한다.
- 변경값을 자유롭게 지정할 수 있다는 장점이 있지만 어떤 데이터들이 있는지 직접 파악해야 한다.
- 데이터를 원래것으로 복원할때는 직접 구현해야 한다.

### LabelEncoder

```python
encoder1 = LabelEncoder()
encoder1.fit(y)
y2 = encoder1.transform(y)
y2
```

```python
y3 = encoder1.inverse_transform(y2)
y3
```

- LabelEncoder
- 컬럼안에 저장되어 있는 값을 확인하지 않아도 되는건 fatorize랑 동일하다.
- 값의 변환 원리도 fatorize와 동일하다.
- 다만, 값을 원래대로 복원하는 함수를 제공한다는 것이 다르다.
