```python
# 기본
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 경고 뜨지 않게...
import warnings
warnings.filterwarnings('ignore')

# 그래프 설정
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['axes.unicode_minus'] = False

# 데이터 전처리 알고리즘
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

# 학습용과 검증용으로 나누는 함수
from sklearn.model_selection import train_test_split

# 교차검증
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold

# 하이퍼 파라미터 튜닝
from sklearn.model_selection import GridSearchCV

# 분류 알고리즘
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# 회귀 알고리즘
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

# 분류용 평가 함수
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

# 회귀용 평가 함수
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

### 앙상블
- 여러 개의 모델을 사용해 결과를 종합해서 결정하는 알고리즘
- 과적합을 이용하여 더 좋은 결과를 낼 수 있도록 고안된 알고리즘

### 운영방식
- 보팅(Voting) : 서로 다른 타입의 모델을 다수 사용하여 같은 데이터를 학습 시키는 방식
- 배깅(Bagging) : 같은 타입의 모델을 다수 사용하여 다른 형태의 데이터를 학습 시키는 방식
- 배깅에서는 데이터를 랜덤하게 섞어서 일부를 샘플링한다.

### 예측방식
- 하드 보팅(HardVoting) : 각 모델이 내린 예측 결과를 분류에서는 다수결, 회귀에서는 평균을 구해 최종 결정한다.
- 소프트 보팅(SoftVoting) : 각 모델이 각각 결과에 대한 확률을 구해 분류에서는 확률의 평균이 가장 높은 것. 회귀에서는 총합을 비율로 설정하여 예측 결과를 계산한다.

### Random Frest
- 다수의 결정트리를 모아 최종 결정을 내리는 모델
- 배깅, 소프트보팅

### 데이터를 불러온다.

```python
df1 = pd.read_csv('data/breast_cancer.csv')
df1
```

### 데이터 전처리

```python
# 입력과 결과로 나눈다.
X = df1.drop('target', axis=1)
y = df1['target']
```

```python
# 문자열 -> 숫자
encoder1 = LabelEncoder()
encoder1.fit(y)
y = encoder1.transform(y)
```

```python
# 표준화
scaler1 = StandardScaler()
scaler1.fit(X)
X = scaler1.transform(X)
```

```python
print(X)
print(y)
```

### 기본 모델을 통한 교차 검증

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=1)
```

```python
model1 = RandomForestClassifier(random_state=1)
r1 = cross_val_score(model1, X, y, scoring='f1', cv=kfold)
print(f'평균 성능 수치 : {r1.mean()}')
```

### 하이퍼 파라미터 튜닝

```python
# 앙상블 모델은 과적합된 트리 모델들을 사용한다.
# 이에, 트리에 관련된 하이퍼 파라미터는 건들지 않는다
# 앙상블 모델 자체의 과적합을 피하기 위해 내부적으로 사용하는
# 트리의 개수를 튜닝한다.
# n_estimators : 내부적으로 사용하는 트리의 개수
params = {
    'n_estimators' : [50, 100, 150, 200, 250, 300]
}

tmp_model1 = RandomForestClassifier(random_state=1)
grid1 = GridSearchCV(tmp_model1, param_grid=params, scoring='f1', cv=kfold)
grid1.fit(X, y)
print(f'최적의 하이퍼 파라미터 : {grid1.best_params_}')
print(f'최적의 모델 평균 성능 : {grid1.best_score_}')
```

```python
print(f'Random Forest Basic : {r1.mean()}')
print(f'Random Forest Tuning : {grid1.best_score_}')
```

### 최종 평가를 수행한다.

```python
best_model1 = grid1.best_estimator_
```

```python
# 학습용과 검증용으로 데이터를 나눈다.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# 학습용 데이터로 학습을 시킨다.
best_model1.fit(X_train, y_train)
```

```python
# 학습용 데이터를 통해 예측 결과를 가져온다.
y_pred1 = best_model1.predict(X_train)
```

```python
r1 = accuracy_score(y_train, y_pred1)
r2 = precision_score(y_train, y_pred1)
r3 = recall_score(y_train, y_pred1)
r4 = f1_score(y_train, y_pred1)
r5 = roc_auc_score(y_train, y_pred1)

print('Random Forest 학습 데이터 평가')
print(f'accuracy : {r1.mean()}')
print(f'precision : {r2.mean()}')
print(f'recall : {r3.mean()}')
print(f'f1 : {r4.mean()}')
print(f'roc_auc : {r5.mean()}')
```

```python
proba1 = best_model1.predict_proba(X_train)

# 0일 확률
a10 = proba1[:, 0]
# 1일 확률
a11 = proba1[:, 1]

plt.scatter(list(range(len(a10))), a10, label='0일 확률')
plt.scatter(list(range(len(a11))), a11, label='1일 확률')
plt.title('Random Forest')
plt.legend()
plt.show()
```

```python
# 예측 결과와 진짜 결과의 패턴을 시각화
plt.scatter(list(range(len(y_train))), y_train, label='original')
plt.scatter(list(range(len(y_pred1))), y_pred1, label='prediction')
plt.title('Random Forest')
plt.legend()
plt.show()
```

```python
# 검증용 데이터를 통해 예측 결과를 가져온다.
y_pred1 = best_model1.predict(X_test)
```

```python
r1 = accuracy_score(y_test, y_pred1)
r2 = precision_score(y_test, y_pred1)
r3 = recall_score(y_test, y_pred1)
r4 = f1_score(y_test, y_pred1)
r5 = roc_auc_score(y_test, y_pred1)

print('Random Forest 검증 데이터 평가')
print(f'accuracy : {r1.mean()}')
print(f'precision : {r2.mean()}')
print(f'recall : {r3.mean()}')
print(f'f1 : {r4.mean()}')
print(f'roc_auc : {r5.mean()}')
```

```python
# 예측 결과와 진짜 결과의 패턴을 시각화
plt.scatter(list(range(len(y_test))), y_test, label='original')
plt.scatter(list(range(len(y_pred1))), y_pred1, label='prediction')
plt.title('Random Forest')
plt.legend()
plt.show()
```
