### 선형 모델

- 회귀 : 결과를 일반화할 수 있는 예측선

```python
# 기본
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 경고 뜨지 않게...
import warnings
warnings.filterwarnings('ignore')

# 그래프 설정
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['axes.unicode_minus'] = False

# 데이터 전처리 알고리즘
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

# 학습용과 검증용으로 나누는 함수
from sklearn.model_selection import train_test_split

# 교차검증
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold

# 하이퍼 파라미터 튜닝
from sklearn.model_selection import GridSearchCV

# 분류 알고리즘
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# 분류용 평가 함수
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

# 회귀용 평가 함수
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

- 데이터를 불러온다.
- 적당한 전처리를 수행한다.
- 기본 모델을 통한 교차 검증을 수행한다.
- 하이퍼 파라미터 튜닝을 수행한다.(교차검증)
- 평가 결과를 보고 우수하다고 판단되는 모델을 선정한다.
- 전체 데이터를 학습 시킨다.
- 예측할 데이터가 있다면 예측을 수행한다.

### 데이터를 불러온다.

```python
df1 = pd.read_csv('data/breast_cancer.csv')
df1
```

### 데이터 전처리

```python
# 입력과 결과로 나눈다.
X = df1.drop('target', axis=1)
y = df1['target']
```

```python
# 문자 -> 숫자
encoder1 = LabelEncoder()
encoder1.fit(y)
y = encoder1.transform(y)
```

```python
# 표준화
scaler1 = StandardScaler()
scaler1.fit(X)
X = scaler1.transform(X)
```

```python
print(X)
print(y)
```

### 기본 모델을 통한 교차 검증

```python
# LogisticRegression
model1 = LogisticRegression()
kfold1 = KFold(n_splits=10, shuffle=True, random_state=1)
r1 = cross_val_score(model1, X, y, scoring='f1', cv=kfold1)
print(f'평균 정확도 : {r1.mean()}')
```

```python
# SVM(SVC)
# probability : SVM 모델은 확률을 구하는 과정이 기본적으로 비활성화 되어 있다.
model2 = SVC(probability=True)
kfold2 = KFold(n_splits=10, shuffle=True, random_state=1)
r2 = cross_val_score(model2, X, y, scoring='f1', cv=kfold2)
print(f'평균 정확도 : {r2.mean()}')
```

### 하이퍼 파라미터 튜닝

```python
# LogisticRegression
# 규제를 설정하여 선이 직선보다 조금 휘어지게 만드는 것이 목적이다.
# penalty : 규제 함수의 종류
# C : 규제의 강도
params = {
    'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
    'C' : [0.0001, 0.001, 0.01, 0.1, 0, 1, 10, 100, 1000]
}

model1 = LogisticRegression()
kfold1 = KFold(n_splits=10, shuffle=True, random_state=1)
grid1 = GridSearchCV(model1, param_grid=params, scoring='f1', cv=kfold1)
grid1.fit(X, y)
print(f'최적의 하이퍼 파라미터 : {grid1.best_params_}')
print(f'최적의 모델 평균 성능 : {grid1.best_score_}')
```

```python
# SVC
# 규제를 설정하여 선이 직선보다 조금 휘어지게 만드는 것이 목적이다.
# C : 규제의 강도
# 규제 함수는 l2로 고정되어 있다.
params = {
    'C' : [0.0001, 0.001, 0.01, 0.1, 0, 1, 10, 100, 1000]
}

model2 = SVC(probability=True)
kfold2 = KFold(n_splits=10, shuffle=True, random_state=1)
grid2 = GridSearchCV(model2, param_grid=params, scoring='f1', cv=kfold2)
grid2.fit(X, y)
print(f'최적의 하이퍼 파라미터 : {grid2.best_params_}')
print(f'최적의 모델 평균 성능 : {grid2.best_score_}')
```

### 최적의 모델에 전체 데이터를 학습한다.

```python
# 학습과 검증으로 나눈다.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# 최적의 모델을 가져온다.
best_model1 = grid1.best_estimator_
best_model2 = grid2.best_estimator_
```

```python
print(best_model1)
print(best_model2)
```

```python
# 학습한다.
best_model1.fit(X_train, y_train)
best_model2.fit(X_train, y_train)
```

```python
# 학습한 데이터를 통한 평가
# 학습한 데이터를 통한 평가이기 때문에 성능은 무조건 잘나와야 한다.
y_pred1 = best_model1.predict(X_train)
y_pred2 = best_model2.predict(X_train)
```

```python
r1 = accuracy_score(y_train, y_pred1)
r2 = precision_score(y_train, y_pred1)
r3 = recall_score(y_train, y_pred1)
r4 = f1_score(y_train, y_pred1)
r5 = roc_auc_score(y_train, y_pred1)

print('LogisticRegression 학습 데이터 평가')
print(f'accuracy : {r1.mean()}')
print(f'precision : {r2.mean()}')
print(f'recall : {r3.mean()}')
print(f'f1 : {r4.mean()}')
print(f'roc_auc : {r5.mean()}')
```

```python
# LogisticRegression에 대한 예측 확률
proba1 = best_model1.predict_proba(X_train)

# 0일 확률
a10 = proba1[:, 0]
# 1일 확률
a11 = proba1[:, 1]

plt.scatter(list(range(len(a10))), a10, label='0일 확률')
plt.scatter(list(range(len(a11))), a11, label='1일 확률')
plt.title('LogisticRegression')
plt.legend()
plt.show()
```

```python
r1 = accuracy_score(y_train, y_pred2)
r2 = precision_score(y_train, y_pred2)
r3 = recall_score(y_train, y_pred2)
r4 = f1_score(y_train, y_pred2)
r5 = roc_auc_score(y_train, y_pred2)

print('SVM(SVC) 학습 데이터 평가')
print(f'accuracy : {r1.mean()}')
print(f'precision : {r2.mean()}')
print(f'recall : {r3.mean()}')
print(f'f1 : {r4.mean()}')
print(f'roc_auc : {r5.mean()}')
```

```python
# SVM(SVC)에 대한 예측 확률
proba1 = best_model2.predict_proba(X_train)

# 0일 확률
a10 = proba1[:, 0]
# 1일 확률
a11 = proba1[:, 1]

plt.scatter(list(range(len(a10))), a10, label='0일 확률')
plt.scatter(list(range(len(a11))), a11, label='1일 확률')
plt.title('SVM(SVC)')
plt.legend()
plt.show()
```

```python
# 검증용 데이터를 통한 평가
# 학승용 데이터 평가 결과보다 너무 낮게 나오면 해당 모델을 과적합현상
# (학습 데이터에 너무 최적화되어 새로운 데이터를 예측하지 못하는 현상)이
# 발생한 것으로 봐야 한다.
y_pred1 = best_model1.predict(X_test)
y_pred2 = best_model2.predict(X_test)
```
