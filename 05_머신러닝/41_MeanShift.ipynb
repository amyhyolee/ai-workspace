### Mean Shift
- 개발자가 정해준 거리를 기반으로 밀집도가 높은 데이터들 끼리 그룹으로 묶는 군집
- 그룹의 개수를 예측이 불가능하며 군집의 수를 줄이려면 반경의 크기를 늘리고 군집의 수를 늘리려면 반경의 크기를 줄인다.
