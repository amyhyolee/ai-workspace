### Label Encoder
- 문자 데이터가 저장되어 있는 컬럼의 데이터를 숫자로 변환하는 작업을 한다.
- 머신러닝 알고리즘 중에서는 문자 데이터를 학습하지 못하는 알고리즘들이 존재한다.
- 이에 모든 알고리즘을 사용해야 하기 때문에 문자 데이터를 숫자로 변환하는 작업을 해야 한다.
- 지정된 컬럼에 저장되어 있는 값들을 확인하고 각 값들에 숫자를 부여해 반환해준다.
- 복원도 가능하다.

```python
# pandas의 map 사용
# 먼저 컬럼에 어떠한 값들이 저장되어 있는지 확인해야 한다.
iris_df['target'].value_counts().index
```

```python
# 확인된 값들을 딕셔너리를 통해 변환될 값을 지정한다.
dict1 = {
    'setosa' : 0,
    'versicolor' : 1,
    'virginica' : 2
}

# map 함수를 이용해 변환한다.
a1 = iris_df['target'].map(dict1)
a1
```

```python
# 복원
dict2 = {
    0 : 'setosa',
    1 : 'versicolor',
    2 : 'virginica'
}

a2 = a1.map(dict2)
a2
```

```python
# LabelEncoder
encoder1 = LabelEncoder()
# 데이터를 학습한다.
encoder1.fit(iris_df['target'])
# 학습한 것을 토대로 변환한다.
a1 = encoder1.transform(iris_df['target'])
a1
```

```python
# 복원
a2 = encoder1.inverse_transform(a1)
a2
```

### 표준화 작업
- 학습 데이터의 각 컬럼의 데이터를 비슷한 수준의 범위로 맞추는 작업
- 성능을 향상 시킬 수 있는 효과를 얻을 수 있다.
- 시각화 할 때도 도움이 된다.
- 주로 공식에 의해서 계산하는 학습 알고리즘에 대해 큰 영향을 미친다.

```python
# 표준화 작업을 위해 결과 데이터를 제외한다.
X = iris_df.drop('target', axis=1)
X
```

```python
# 표준화
scaler1 = StandardScaler()
scaler1.fit(X)
scaler1_df = scaler1.transform(X)
scaler1_df
```
