### 사용 모듈
- requests 모듈 : 직접 주소를 요청할 수 있다. requests 모듈을 통해 요청해서 받아오는 데이터는 웹 브라우저에서 소스보기 했을 때 나오는 코드이다. 따라서 requests 모듈을 통해 받아온 데이터에는 브라우저 화면에 있는 데이터가 없을 수도 있다.
- selenium 모듈 : 웹 브라우저를 조작하여 원하는 페이지로 이동한 다음에 데이터를 받아온다. 크롬 개발자 도구의 elements 탭에 있는 데이터를 받아 올 수 있다.
- bs4 : html, xml 데이터를 분석하는 용도
- json : json 데이터를 분석하는 용도
- feedparser : rss 데이터를 분석하는 용도

### 웹 문서 수집시 알아야할 사항
- 주소 분석이 가능하다면 requests 모듈로 요청한다.
- 주소 분석이 어렵다면 selenium 모듈을 사용한다.
- requests 모듈을 속도가 빠르고 selenium은 속도가 느리다.

- 소스보기 했을 때 데이터가 있다면 : requests 모듈로 요청하고 bs4로 수집한다.
- 소스보기 했을 때 데이터가 없고 elements 탭에 데이터가 있는 경우 : ajax인가 아닌가...
    - ajax인 경우 : 요청 주소를 파악해서 requests로 요청한다.
    - iframe인 경우 : iframe 태그의 src 속성의 주소를 파악하여 requests 모듈로 요청한다.
    - 웹소켓인 경우 : selenium으로 데이터를 수집한다.
- 소스보기 했을 때 데이터가 없고 elements 탭에도 데이터가 없다면 iframe 여부를 확인한다.
- iframe이 있는지 확인해 보고 있다면 src 속성의 주소를 통해 요청한다. (iframe: 문서안에 문서 넣기)
- 위의 경우가 모두 아니라면 selenium을 통해 시도해본다.
- 위의 경우가 모두 어렵다고 생각이 되면 selenium을 사용한다.


- iframe: 프레임소스보기로 알 수 있음
- ajax: Network 탭 (xhr) --> 사이트에서 원하는 버튼? 누르고 Name 더블클릭
- 웹소켓: Network 탭 (Fetch) --> Selenium 사용

1. 소스보기 했을 때 소스에 데이터가 있다 
   -> 주소 창의 주소를 이용해 requests 요청, bs4 분석
2. 소스보기 했을 때 없고 마우스 우클릭 메뉴에 프레임 소스보기가 있다.
   -> frame 태그의 src 속성의 주소를 이용해 requests 요청, bs4 분석
3. 소스보기 했을 때 없고 마우스 우클릭 메뉴에 프레임 소스보기가 없다.
   -> 개발자 도구의 elements 탭에서 데이터가 있는지 확인.
4. 데이터가 있다면 ajax인지 웹 소켓인지 .. .확인
5. 개발자 도구의 network 탭에서 페이지가 변경되는 작업을 했을 때
   xhr 이라고 나오면 ajax. 이 주소를 이용해 requests 요청, bs4 분석
6. fetch로 나오면 웺 소켓 -> 셀레니움을 이용해 분석

### 태그를 지칭하기 위해 알아야 할 것들.
- 원하는 데이터를 가지고 있는 태그를 지칭할 수 있어야 필요한 데이터를 수집할 수 있다.
- id 속성 : html에서 사용하는 모든 태그에 사용할 수 있는 속성. 이 속성은 하나의 HTML 문서 안에 같은 id를 가지고 있는 태그는 없다. id는 웹 개발자가 특정 태그 하나를 지칭하기 위해 사용한다. 우리는 데이터를 가지고 있는 특정 태그를 지칭하는 용도로 사용할 수 있다.
- class 속성 : css는 html 태그를 꾸미는 용도로 사용한다. 이러한 css들을 미리 정의해놓고 태그에 적용할 때 사용하는 것이 class 속성이다. 웹을 개발할 때 특정 영역의 모양을 설정하기 위해 그 영역만을 위한 css를 만들어 사용하는 경우가 많다. 만약 우리가 필요로 하는 데이터가 특정 영역에 있고 그 영역에서만 사용하는 class 속성이 있다면 이를 통해 접근할 수 있다.
- nth-child : 같은 태그가 동시에 여러개 내열되어 있을 경우 그 순서를 구분하기 위해 사용하는 것(1부터 시작)


### 코드 작성 방법
- 한 페이지의 데이터를 수집한다.
- 다음 페이지로 넘어갈 수 있는 방법을 찾는다.
- 마지막 페이지 여부를 어떻게 확인할 수 있는지 찾는다.
- 반복문을 통해 마지막 페이지까지 갈 수 있도록 구성해준다.
- 반드시 수집을 위해 만들 프로그램이 종료가 되게 만들어야 한다.
- 반드시 요청 전에 딜레이를 줘야 한다(최장 3초)
