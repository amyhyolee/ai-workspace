- selenium 사용을 위해 webdriver를 구해야 한다.
- web driver는 브라우저 제조사들이 제공하기 때문에 브라우저 제조사에서 내려받아야 한다.
- webdriver로 검색
- https://chromedriver.chromium.org/downloads

- 우측 상단의 쩜쩜쩜(메뉴)를 클릭한다. 
- 도움말 > 크롬 정보에서 버전을 확인한다.

```python
from selenium import webdriver
```

```python
# 웹 드라이버 가동
driver = webdriver.Chrome('chromedriver')
```

```python
# 구글 요청
driver.get('https://google.com')
```

```python
# 네이버 요청
driver.get('https://naver.com')
```
