### 셀
- 설명이나 코드를 작성할 수 있는 칸 하나를 의미한다.

### 셀 모드
- command mode(esc) : 주피터에 있는 셀들을 관리할 수 있다.
- edit mode(enter) : 현재 셀에 대해 편집 작업을 할 수 있다.

### 공통
- shift + enter : 현재 셀을 실행한다. 실행 후 다음 셀로 이동.
- ctrl + enter : 현재 셀을 실행한다. 다음 셀로 이동하지 않는다.

### command mode
- 상하방향키 : 셀 이동
- a : 현재 셀의 상단에 셀을 추가
- b : 현재 셀의 하단에 셀을 추가
- dd : 현재 셀을 삭제한다.
- m : 현재 셀을 markdown(설명작성)로 변경한다.
- y : 현재 셀을 코드 작성 타입으로 변경한다.
- shift + m : 선택한 셀들을 하나의 셀로 합친다.
- z : 삭제한 셀을 복원한다.

### edit mode
- 상하 방향키 : 현재 셀 상단에 있을 경우 이전셀로 이동하고 제일 아래에 있을 경우에는 다음 셀로 이동한다.
- ctrl + shift + - : 현재 위치를 기준으로 셀을 나눈다.
