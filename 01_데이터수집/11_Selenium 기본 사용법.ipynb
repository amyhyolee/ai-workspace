## Selenium 기본 사용법

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
````

````python
# 웹 드라이버 실행
driver = webdriver.Chrome('chromedriver')
````

````python
# 네이버를 요청한다.
driver.get('https://naver.com')
````

````python
# 회원 가입 링크를 누른다.
time.sleep(1)

join_tag = driver.find_element(By.CSS_SELECTOR, '#account > div > a')
join_tag.click()
````

```python
# 위에 코드랑 같지만 deprecated 없어질수도 있는 코드
# 회원 가입 링크를 누른다.
time.sleep(1)

join_tag = driver.find_element_by_css_selector('#account > div > a')
join_tag.click()
```

````python
# 약관 전체 동의를 클릭한다.
check_tag = driver.find_element(By.CSS_SELECTOR, '#join_form > div.terms_p > p > span > label')
check_tag.click()
````

````python
# 약관의 확인 버튼을 누른다.
time.sleep(1)

agree_btn = driver.find_element(By.CSS_SELECTOR, '#btnAgree')
agree_btn.click()
````
