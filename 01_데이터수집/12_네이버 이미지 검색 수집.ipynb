## [네이버 이미지 검색 수집](http://www.naver.com)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
from IPython.display import clear_output
import requests
```

```python
# 검색어를 입력받는다.
keyword = input('검색어를 입력해주세요 : ')
```

```python
# 웹 드라이버 실행
driver = webdriver.Chrome('chromedriver')
```

```python
# 네이버 사이트를 요청한다.
time.sleep(1)
driver.get('https://naver.com')
```

```python
# 검색어를 입력한다.
time.sleep(1)

input_tag = driver.find_element(By.CSS_SELECTOR, '#query')
input_tag.send_keys(keyword)
```

```python
# 검색 버튼을 누른다.
time.sleep(1)

search_btn = driver.find_element(By.CSS_SELECTOR, '#search_btn')
search_btn.click()
```

```python
# 이미지 탭을 누른다.
time.sleep(1)

# 두 번째 탭을 눌러을 눌러준다.
# image_tab = driver.find_element(By.CSS_SELECTOR, '#lnb > div.lnb_group > div > ul > li:nth-child(2) > a')
# image_tab.click()

# 검색어 따라 순서가 바뀔 수도 있기 때문에 "이미지"라는 a태그를 누르도록 한다.
image_tab = driver.find_element(By.LINK_TEXT, '이미지')
image_tab.click()

# 만약 a 태그가 아니라면 아래 코드처럼 직접 찾아줘야 한다.
# tab_tags = driver.find_elements(By.CSS_SELECTOR, '#lnb > div.lnb_group > div > ul > li > a')
# for tab_tag in tab_tags :
#     if tab_tag.text.strip() == '이미지' :
#        tab_tag.click()
```

```python
# 스크롤 한다.
for idx in range(10) :
    time.sleep(1)
    # 스크롤 할 수 있는 javascript 코드를 실행 시킨다.
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
```

```python
# 이미지들이 있는 태그들을 가져온다.
imgs = driver.find_elements(By.CSS_SELECTOR, '#main_pack > section.sc_new.sp_nimage._prs_img._imageSearchPC > div > div.photo_group._listGrid > div.photo_tile._grid > div > div > div.thumb > a > img')

image_list = []

for img in imgs :
    # print(img)
    src = img.get_attribute('src')
    # print(src)
    # http로 시작하는 것만 사용한다
    if src.startswith('http') :
        # 이미지의 주소를 리스트에 담는다.
        image_list.append(src)
```

```python
# 이미지를 저장한다.
for idx, src in enumerate(image_list) :
    time.sleep(0.5)
    clear_output(wait=True)
    
    print(f'{idx + 1} / {len(image_list)}')
    
    # 이미지 데이터를 받는다.
    img_res = requests.get(src)
    
    # 저장한다.
    with open(f'image/{keyword}{idx}.jpg', 'wb') as fp :
        fp.write(img_res.content)
        
print('저장완료')
```
