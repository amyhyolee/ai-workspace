## 셀레니움 네이버 영화

### 셀레니움 사용

1. https://naver.com 으로 들어가고

2. 검색어에 "네이버 영화"로 입력하고 검색을 한다.

3. 검색 결과에서 네이버 영화 링크를 클릭하고 들어간다.

4. 좌측의 "상영작.예정작"을 클릭한다.

5. 나오는 영화목록들의 영화제목, 평점, 예매율을 수집해서 csv로 저장한다.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
from IPython.display import clear_output
import requests

import numpy as np
import pandas as pd
import os
```

```python
# 웹 드라이버 실행
driver = webdriver.Chrome('./수업중_사용한_프로그램/chromedriver')
```

```python
# 네이버 요청
driver.get('https://naver.com')
```

```python
# 검색어 입력
a1 = driver.find_element(By.CSS_SELECTOR, '#query')
a1.send_keys('네이버 영화')
```

```python
# 검색 버튼 클릭
a1 = driver.find_element(By.CSS_SELECTOR, '#search_btn')
a1.click()
```

```python
# 검색 결과에서 네이버 영화를 클릭한다.
a1 = driver.find_element(By.CSS_SELECTOR, '#main_pack > section.sc_new.sp_nsite._project_site_channel_root._section_nsite_._sp_ntotal._prs_vsd_bas > div > div > div.nsite_tit > div > div.nsite_name > a')
a1.click()
```

```python
# 상영 예정작 클릭
# 탭들을 가져온다.
tabs = driver.window_handles
# 두 번째 탭을 제어하도록 설정한다.
driver.switch_to.window(tabs[1])

# 상영작,예정작을 클릭한다.
a1 = driver.find_element(By.CSS_SELECTOR, '#scrollbar > div.scrollbar-box > div > div > ul > li:nth-child(2) > a')
a1.click()
```

```python
# 영화들을 가져온다.
li_list = driver.find_elements(By.CSS_SELECTOR, '#content > div.article > div:nth-child(1) > div.lst_wrap > ul > li')

data_dict = {
    '제목' : [],
    '평점' : [],
    '예매율' : []
}

# 영화의 수 만큼 반복한다.
for li in li_list : 
    # 제목
    a1 = li.find_element(By.CSS_SELECTOR, 'dl > dt > a')
    data1 = a1.text.strip()
    
    # 평점
    a1 = li.find_element(By.CSS_SELECTOR, 'dl > dd.star > dl.info_star > dd > div > a > span.num')
    data2 = a1.text.strip()
    
    # 예매율 
    a1 = li.find_element(By.CSS_SELECTOR, 'dl > dd.star > dl.info_exp > dd > div > span.num')
    data3 = a1.text.strip()
    
    # print(data1)
    # print(data2)
    # print(data3)
    # print('-------------------------------')
    data_dict['제목'].append(data1)
    data_dict['평점'].append(data2)
    data_dict['예매율'].append(data3)
```

```python
# 데이터 프레임 생성
df1 = pd.DataFrame(data_dict)

fileName = 'selenium_naver_movie.csv'

# 파일이 존재하면 삭제한다.
if os.path.exists(fileName) == True :
    os.remove(fileName)

# 저장한다.
df1.to_csv(fileName, encoding='utf-8-sig', index=False)
```
