### 링크 주소 정보를 가져온다.

```python
link_df = pd.read_csv('data/namu_wiki_link.csv')
link_df
```

### 데이터를 수집한다.

```python
# 데이터를 담을 딕셔너리
data_dict = {
    'title' : [],
    'text' : []
}

# 주소의 수 만큼 반복한다.
for idx, title in enumerate(link_df['title']) :
    # 1초 쉰다.
    time.sleep(1)
    clear_output(wait=True)
    
    # 링크 주소
    link = link_df['link'][idx]
    
    print(title)
    print(link)
    
    data_dict['title'].append(title)
    
    # 브라우저를 실행시킨다(캡차 때문에 매번 새롭게 실행시킨다)
    driver = webdriver.Chrome('./chromedriver')
    driver.get(link)
    time.sleep(1)

    # 문자열 내용만 모두 가져온다.
    a1 = driver.find_elements_by_css_selector('div.wiki-paragraph')
    
    text2 = ''
    for a2 in a1 :
        text2 = text2 + a2.text + ' '
    
    # print(text2)
    
    data_dict['text'].append(text2.strip())
    
# 저장한다.   
df1 = pd.DataFrame(data_dict)
# 저장한다.
# 파일이 없다면 : 첫 번째 저장
if os.path.exists(fileName) == False :
    df1.to_csv(fileName, encoding='utf-8-sig', index=False)
# 파일이 있다면..
else :
    df1.to_csv(fileName, encoding='utf-8-sig', index=False, mode='a', header=None)
    
print('저장완료')
```
