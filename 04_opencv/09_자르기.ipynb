```python
import cv2
```

```python
# 이미지를 가져온다.
src = cv2.imread('opencv_image/7.jpg', cv2.IMREAD_REDUCED_COLOR_2)
```

```python
# 이미지의 일부만 가져온다.
# 행렬 특성상 (세로, 가로)가 된다.
# 이미지 행렬 데이터의 일부분을 가져와 그린다.
dst = src[100:600, 200:700]
```

```python
cv2.imshow('src', src)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```
