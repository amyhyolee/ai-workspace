- Haar Classifier 사용을 위한 케스케이드 파일
- Haar Object Dectection은 학습을 통해 구해진 가중치와 레이어의 구조가 정의되어 있는 파일을 불러와 사물을 인식하는 형태로 동작을 한다.
- 학습 과정이라는 것은 존재하지 않으나, 작업속도가 빠르지만 인식하지 못하는 사물이 존재할 수 있다.

```python
# cascade file
face_cascade_name = 'opencv_data/haarcascades/haarcascade_frontalface_alt.xml'
eye_cascade_name = 'opencv_data/haarcascades/haarcascade_eye_tree_eyeglasses.xml'
```

```python
# 학습 알고리즘을 생성한다.
face_cascade = cv2.CascadeClassifier()
eyes_cascade = cv2.CascadeClassifier()
```

```python
# 각 학습 알고리즘에 cascade 파일을 지정한다.
face_cascade.load(cv2.samples.findFile(face_cascade_name))
eyes_cascade.load(cv2.samples.findFile(eye_cascade_name))
```

```python
# 사용할 이미지를 그레이 스케일로 변환한다(인식률 향상)
frame_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 이미지데이터를 표준화한다(인식률 향상)
frame_gray = cv2.equalizeHist(frame_gray)
```

```python
# 얼굴을 인식한다.
# 인식된 부분은 이미지상의 좌측 상단 지점 x, y 좌표와 가로, 세로 길이를 
# 가져온다.
faces = face_cascade.detectMultiScale(frame_gray)
faces
```

```python
# 찾은 얼굴의 수 만큼 반복한다.
for x, y, w, h in faces :
    # 인식한 얼굴에 네모를 그려준다.
    # 이미지, 좌측상단 x, y, 우측하단 x, y, 색상, 두께
    frame = cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
    
    # 눈을 찾기 위해 얼굴 데이터만 가져온다.
    faceROI = frame_gray[y:y + h, x:x + w]
    # 눈을 검출한다.
    eyes = eyes_cascade.detectMultiScale(faceROI)
    
    # print(eyes)        
    
    # 검출한 눈의 수만큼 반복하여 동그라미를 그린다.
    for x2, y2, w2, h2 in eyes :
        # 눈의 중앙점
        eye_center = x + x2 + w2 // 2, y + y2 + h2 // 2
        # 반경을 계산한다.
        radius = int(round((w2 + h2) * 0.25))
        # 원을 그린다.
        frame = cv2.circle(img, eye_center, radius, (255, 0, 0), 4)
    
    
cv2.imshow('face Detection', frame)
cv2.waitKey()
cv2.destroyAllWindows()
```
