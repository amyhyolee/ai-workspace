```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# opencv
import cv2

# 그래프에 대한 기본설정
# plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 20, 10
plt.rcParams['font.size'] = 16
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 영상 파일을 설정한다.
video = cv2.VideoCapture('opencv_image/video.mp4')
video
```

```python
# 현재 재생되고 있는 프레임 위치
video.get(cv2.CAP_PROP_POS_FRAMES)
```

```python
# 전체 프레임 수
video.get(cv2.CAP_PROP_FRAME_COUNT)
```

```python
# 엔터키를 누를 때 까지 반복한다.
while cv2.waitKey(33) < 0 :
    # 현재 재생위치
    a1 = video.get(cv2.CAP_PROP_POS_FRAMES)
    # 전체 프레임 수
    a2 = video.get(cv2.CAP_PROP_FRAME_COUNT)
    
    # 현재 재생 위치가 전체 프레임수와 같다면 처음으로 이동시킨다.
    # 반복재생 구현
    if a1 == a2 :
        video.set(cv2.CAP_PROP_POS_FRAMES, 0)
        
    # 재생
    # 현재 프레임을 추출한다.
    ret, frame = video.read()
    # 현재 프레임을 그린다.
    cv2.imshow('VideoFrame', frame)
    
# 비디오 종료
video.release()
cv2.destroyAllWindows()
```

### VideoCaputre 제공 메서드
- isOpened : 동영상 파일 열기 성공 여부 확인
- open(파일명) : 동영상 파일 열기
- set(속성, 값) : 동영상 속성 설정
- get(속성) : 동영상 속성값 가져오기
- release : 동영상 파일을 닫고 메모리에서 제거

### 속성
- cv2.CAP_PROP_FRAME_WIDTH : 영상 가로길이
- cv2.CAP_PROP_FRAME_HEIGHT : 영상 세로길이
- cv2.CAP_PROP_FRAME_COUNT : 총 프레임수
- cv2.CAP_PROP_FPS : 초당 프레임 수
- cv2.CAP_PROP_FOURCC : 코덱종류
- cv2.CAP_PROP_POS_MESC : 재생시간(MS)
- cv2.CAP_PROP_POS_FRAMES : 재생프레임 위치
- cv2.CAP_PROP_AVI_RATIO : 비디오 파일 상대 위치(0 : 시작, 1 : 끝)
