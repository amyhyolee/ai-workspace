- pyrUp : 피라미드 방식으로 확대한다(4배 확대)
- pyrDown : 피라미드 방식으로 축소한다(1/4로 축소)

```python
import cv2
```

```python
src = cv2.imread('opencv_image/5.jpg', cv2.IMREAD_REDUCED_COLOR_2)
```

```python
# 확대 축소
dist1 = cv2.pyrUp(src)
dist2 = cv2.pyrDown(src)
```

```python
cv2.imshow('src', src)
cv2.imshow('dist1', dist1)
cv2.imshow('dist2', dist2)

cv2.waitKey()
cv2.destroyAllWindows()
```
