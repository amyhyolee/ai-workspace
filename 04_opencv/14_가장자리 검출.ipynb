```python
import cv2
```

```python
src = cv2.imread('opencv_image/12.jpg', cv2.IMREAD_REDUCED_COLOR_2)
```

```python
# 가장자리 부분만 가져온다.
# 100 : 임계값1, 255 : 임계값2
# 임계값1 이하에 해당하는 부분은 모두 제거한다.
# 임계값2 이하에 해당하는 부분은 가장자리고 간주한다.
canny = cv2.Canny(src, 100, 255)
```

```python
cv2.imshow('src', src)
cv2.imshow('canny', canny)

cv2.waitKey()
cv2.destroyAllWindows()
```
