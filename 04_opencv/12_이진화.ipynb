이진화는 흰색, 검정색으로 바꾸는 것을 의미한다.

```python
import cv2
```

```python
src = cv2.imread('opencv_image/10.jpg', cv2.IMREAD_REDUCED_COLOR_2)
```

```python
# 이진화를 위해 그레이스케일로 변환한다.
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
```

```python
# 이진화된 이미지를 가져온다.
# 두 번째 : 임계값
# 세 번째 임계값 보다 클경우 셋팅될 값(255로 설정하면 흰색)
# 색상값이 주어진 임계값 작으면 0(검정)으로 설정하고, 이상이면
# 3번째로 지정한 값을 설정한다.
ret, dst = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
