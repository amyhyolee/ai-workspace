```python
import cv2
```

```python
src = cv2.imread('opencv_image/11.jpg', cv2.IMREAD_REDUCED_COLOR_2)
```

```python
# 블러 처리한다.
# (9, 9) : 흐림 처리를 위한 연산 단위 크기
# 클수록 흐림 정도가 더해진다.
dst = cv2.blur(src, (9, 9))
```

```python
cv2.imshow('src', src)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```
