- 회전은 선형 변환 중 하나이며 회전 변환 행렬을 통해 변환이 진행된다.
- 원하는 각도로 회전할 수 있는 변환 행렬을 구한다음 이 행렬을 이미지에 적용한다.

```python
import cv2
```

```python
# 이미지를 읽어온다.
src = cv2.imread('opencv_image/4.jpg', cv2.IMREAD_REDUCED_COLOR_2)
```

```python
# 이미지에 대한 크기 정보를 가져온다.
width, height, channel = src.shape
```

```python
# 원하는 지점을 기준으로 원하는 만큼 회전시킬 수 있단 변환행렬을 구한다.
center_x = width / 2
center_y = height / 2

# (center_x, center_y) : 회전 시킬 점 위치
# 90 : 회전 각도
# 1 : 변환 크기 비율
matrix = cv2.getRotationMatrix2D((center_x, center_y), 90, 1)
```

```python
# 원본 이미지에 변환 행렬을 적용한다.
dist = cv2.warpAffine(src, matrix, (width, height))
```

```python
cv2.imshow('src', src)
cv2.imshow('dist', dist)

cv2.waitKey()
cv2.destroyAllWindows()
```
