```python
import cv2
```

```python
# 이미지를 불러온다.
src1 = cv2.imread('opencv_image/9.jpg', cv2.IMREAD_REDUCED_COLOR_2)
src2 = cv2.imread('opencv_image/10.jpg', cv2.IMREAD_REDUCED_COLOR_2)
```

```python
# mask 영역에서 서로 공통으로 겹치는 부분 출력
dst1 = cv2.bitwise_and(src1, src2) 
# mask 영역의 두 이미지를 합쳐버림
dst2 = cv2.bitwise_or(src1, src2)
dst3 = cv2.bitwise_xor(src1, src2)
dst4 = cv2.bitwise_not(src1)
```

```python
cv2.imshow('src', src1)
cv2.imshow('and', dst1)
cv2.imshow('or', dst2)
cv2.imshow('xor', dst3)
cv2.imshow('not', dst4)

cv2.waitKey()
cv2.destroyAllWindows()
```
